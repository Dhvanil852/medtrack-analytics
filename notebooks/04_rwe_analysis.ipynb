# 04. Real-World Evidence (RWE) Analysis via DuckDB
Running SQL queries over in-memory pandas DataFrames to answer feasibility questions.

```python
import sys
import os
sys.path.append(os.path.abspath('..'))
import pandas as pd
import duckdb
from src.preprocess import process_pipeline
from src.dataset_generator import generate_synthetic_data

df = process_pipeline(generate_synthetic_data(1000))
conn = duckdb.connect(':memory:')
conn.register('patients', df)
print('DuckDB memory database initialized.')
```

### Question: How many patients aged 50-70 have HbA1c > 7.0?

```python
query1 = """
SELECT COUNT(*) as patient_count
FROM patients
WHERE age_group = '50-70' AND hba1c > 7.0;
"""
conn.execute(query1).fetchdf()
```

### Question: What's the average device reading per diagnosis group?

```python
query2 = """
SELECT diagnosis, ROUND(AVG(device_reading), 2) as avg_device_reading
FROM patients
GROUP BY diagnosis
ORDER BY avg_device_reading DESC;
"""
conn.execute(query2).fetchdf()
```
